```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = '../data/processed/merged_data_2019_2022.csv'
data = pd.read_csv(file_path)
```

```python
# Afficher les premières lignes pour un aperçu
print("Aperçu des données :")
print(data.head())

# Examiner les informations générales sur le dataset
print("\nInformations sur le dataset :")
print(data.info())

# Vérifier les valeurs manquantes
print("\nValeurs manquantes par colonne :")
print(data.isnull().sum())

# Examiner les statistiques descriptives de base
print("\nStatistiques descriptives :")
print(data.describe())

# Examiner la distribution de la variable cible 'grav'
print("\nDistribution de la variable cible 'grav' :")
print(data['grav'].value_counts())
```

```python
from matplotlib.colors import LinearSegmentedColormap

# Distribution globale des classes de 'grav'
grav_distribution = data['grav'].value_counts(normalize=True)

# Identifier les colonnes catégorielles
categorical_features = data.select_dtypes(include=['object', 'category']).columns

# Créer une colormap personnalisée
custom_cmap = LinearSegmentedColormap.from_list("custom_cmap", ["skyblue", "white", "salmon"])

# Fixer des limites cohérentes pour la colormap
vmin, vmax = 80, 120

# Calcul des pourcentages ajustés
for feature in categorical_features:
    # Table de fréquences croisées normalisée par ligne
    crosstab = pd.crosstab(data[feature], data['grav'], normalize='index') * 100

    # Ajustement selon la distribution globale de 'grav'
    for col in crosstab.columns:
        crosstab[col] = crosstab[col] / grav_distribution[col]
    
    print(f"\nRapport ajusté pour la variable '{feature}' :")
    print(crosstab)
    
    # Visualisation sous forme de heatmap avec limites fixes
    plt.figure(figsize=(10, 8))
    sns.heatmap(crosstab, annot=True, fmt=".2f", cmap=custom_cmap, vmin=vmin, vmax=vmax)
    plt.title(f'Heatmap des rapports ajustés entre {feature} et grav', fontsize=14)
    plt.xlabel('Catégories de grav', fontsize=12)
    plt.ylabel(feature, fontsize=12)
    plt.show()
```

```python
# Filtrer les données pour grav == 2
reg_grav_2 = data[data['grav'] == 2]['reg']

# Calculer les occurrences de chaque catégorie dans 'reg' pour grav == 2
reg_counts = reg_grav_2.value_counts()
    
# Visualiser la distribution avec un barplot
plt.figure(figsize=(12, 8))  # Augmenter la taille pour plus de lisibilité
reg_counts.plot(kind='bar', color='salmon', edgecolor='black')
plt.title('Distribution de "reg" pour grav = 2', fontsize=14)
plt.xlabel('Catégories de reg', fontsize=12)
plt.ylabel('Nombre d\'occurrences', fontsize=12)

# Rotation des labels pour une meilleure lisibilité
plt.xticks(rotation=30, ha='right', fontsize=10)  # Inclinaison et alignement des labels
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()  # Ajuster les marges pour éviter que les labels soient coupés
plt.show()
```

```python
# Calculer les occurrences totales de chaque région
total_counts = data['reg'].value_counts()

# Calculer les occurrences de grav = 2 pour chaque région
grav_2_counts = data[data['grav'] == 2]['reg'].value_counts()

# Calculer la proportion de grav = 2 par rapport au total pour chaque région
proportions = (grav_2_counts / total_counts) * 100

# Remplacer les NaN par 0 pour les régions sans grav = 2
proportions = proportions.fillna(0)

# Calculer la moyenne des proportions
median_proportion = proportions.median()

# Visualiser la proportion avec un barplot
plt.figure(figsize=(12, 8))
proportions.sort_values(ascending=False).plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Proportion de grav = 2 par rapport au total pour chaque région', fontsize=14)
plt.xlabel('Régions', fontsize=12)
plt.ylabel('Proportion (%)', fontsize=12)

# Ajouter une ligne pointillée rouge pour la moyenne
plt.axhline(y=median_proportion, color='red', linestyle='--', linewidth=1.5, label=f'Mediane = {median_proportion:.2f}%')

# Rotation des labels pour une meilleure lisibilité
plt.xticks(rotation=30, ha='right', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Ajouter une légende
plt.legend(loc='upper right', fontsize=12)
plt.tight_layout()  # Ajuster les marges pour éviter que les labels soient coupés
plt.show()
```

```python
# Identifier les colonnes catégoriques
categorical_features = data.select_dtypes(include=['object', 'category']).columns

# Générer un stacked barplot pour chaque variable catégorielle
for feature in categorical_features:
    # Vérifier si la colonne existe et n'est pas vide
    if feature in data.columns and not data[feature].isnull().all():
        # Calculer les proportions de grav 1 et grav 2 pour chaque catégorie
        crosstab = pd.crosstab(data[feature], data['grav'], normalize='index') * 100

        # Trier les catégories par proportion de grav = 2 (ordre ascendant)
        crosstab = crosstab.sort_values(by=2)  # Trier par la colonne correspondant à grav = 2

        # Créer un stacked barplot
        crosstab.plot(
            kind='bar',
            stacked=True,
            color=['skyblue', 'salmon'],
            edgecolor='black',
            figsize=(10, 6)
        )

        # Personnaliser le graphique
        plt.title(f'Distribution des catégories de "{feature}" par gravité (en proportion)', fontsize=16, fontweight='light')
        plt.xlabel(f'Catégories de {feature}', fontsize=14)
        plt.ylabel('Proportion (%)', fontsize=14)
        plt.xticks(rotation=30, ha='right', fontsize=12)
        plt.yticks(fontsize=12)
        plt.legend(['Grav = 1', 'Grav = 2'], title='Gravité', fontsize=12, title_fontsize=14)
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.tight_layout()  # Ajuster les marges
        plt.show()
```
